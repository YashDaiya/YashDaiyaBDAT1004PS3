```python
#Reading CSV file and store it in DataFrame
Data = pd.read_csv('./us-marriages-divorces.csv')
Data
```

```python
import matplotlib.pyplot as plt           #Importing Ploting Library

#Line1 X and Y
x1 = Data['Year']
y1 = Data['Marriages']

#Ploting First data
plt.plot(x1, y1, label = "Marriages")

#Line2 X and Y 
x2 = Data['Year']
y2 = Data['Divorces']

#ploting Second Data
plt.plot(x2, y2, label = "Divorces")

plt.xlabel('Years')                       #XAxes label
plt.ylabel('Counts length-6')             #YAxes label- it's not showing whole number but indicates that it is 6 in the length.
plt.title('Marriages & Divorces in the U.S. between 1867-2014')
plt.legend()                              #Showing Legends
plt.show()                            
```

```python
import numpy as np  #importing numpy
import pandas as pd #importing Pandas   

Data = pd.read_csv('./us-marriages-divorces.csv')


bar1= Data[Data.Year == 1900]               #Filtering data for 1900

bar2 = Data[Data.Year == 1950]              #Filtering data for 1950

bar3 = Data[Data.Year == 2000]              #Filtering Data for 2000

import matplotlib.pyplot as plt             #Importing ploting library

x1 = bar1.Marriages_per_1000
y1 = bar1.Divorces_per_1000
plt.bar(x1, y1, label = "Year 1900")        #Ploting first Bar

x2= bar2.Marriages_per_1000
y2 = bar2.Divorces_per_1000
plt.bar(x2, y2, label = "Year 1950")        #Ploting Second Bar

x3= bar3.Marriages_per_1000
y3 = bar3.Divorces_per_1000
plt.bar(x3, y3, label = "Year 2000")        #Ploting Third Bar

plt.title('Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
plt.legend()
plt.show()
```
